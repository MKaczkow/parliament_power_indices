## Example

```python
game = px.Game(
    quota=simple_majority_threshold,
    weights=[
        law_and_justice['seats'],
        sovereign_poland['seats'],
        the_republicans['seats'],
        reneval['seats'],
        polish_affairs['seats'],
        civic_platform['seats'],
        modern['seats'],
        polish_initiative['seats'],
        the_greens['seats'],
        new_left['seats'],
        left_together['seats'],
        polish_peoples_party['seats'],
        centre_for_poland['seats'],
        union_of_eu_democrats['seats'],
        national_movement['seats'],
        new_hope['seats'],
        confederation_polish_crown['seats'],
        poland_2050['seats'],
        agreement['seats'],
        wolnosciowcy['seats'],
        kukiz_15['seats'],
        democratic_left_assoc['seats']
    ]
)

game.calc()
print("Banzhaf index:")
print(game.banzhaf)
print("Shapley-Shubik index:")
print(game.shapley_shubik)
```

```python
for i, value in enumerate(game.banzhaf):
    parties_list[i]['banzhaf_power'] = value
```

```python
for i, value in enumerate(game.shapley_shubik):
    parties_list[i]['shapley_shubik_power'] = value
```

```python
print(parties_list)
```

```python
df = pd.DataFrame(parties_list)
df
```
